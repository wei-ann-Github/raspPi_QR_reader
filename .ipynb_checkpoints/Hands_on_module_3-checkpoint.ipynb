```python
NAME = ['ZhouXin', 'YiDeliang', 'WangYongjie', 'ZhangHuaizheng', 'LuoYong']
datagen = ImageDataGenerator(
        rotation_range=90,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        rescale=0.5,
        fill_mode='nearest')
  
# create folder to store more training data
if not os.path.isdir('aug_data'):
  os.mkdir('aug_data')
if not os.path.isdir('aug_data/train'):
  os.mkdir('aug_data/train')
for i in NAME:
  if not os.path.isdir('aug_data/train/' + i):
    os.mkdir('aug_data/train/' + i)

# the .flow() command below generates batches of randomly transformed images
# and saves the results to the `aug_data/` directory
for name in NAME:
    for number in range(1, 6):
        temp = load_img('face_data/' + name + '/' + name + '00' + str(number) + '.jpeg')
        img_arr = img_to_array(temp)
        img_arr = img_arr.reshape((1, ) + img_arr.shape)
        i = 0
        for batch in datagen.flow(img_arr, batch_size=1,
                                  save_to_dir='aug_data/train/'+ name, save_prefix=name + str(number), save_format='jpeg'):
            i += 1
            print 'Now is {}th {}'.format(i, name)
            if i > 99:
                break
```

```python
def model(input_shape):
    X_input = Input(input_shape)
    X = Conv2D(32, (3, 3), )(X_input)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = MaxPooling2D((2, 2))(X)
    
    X = Conv2D(32, (3, 3), )(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = MaxPooling2D((2, 2))(X)
    
    X = Conv2D(64, (3, 3), )(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = MaxPooling2D((2, 2))(X)
    
    X = Flatten()(X)
    X = Dense(64)(X)
    X = Activation('relu')(X)
    X = Dropout(0.5)(X)
    X = Dense(5)(X)
    X = Activation('softmax')(X)
    
    model = Model(inputs=X_input, outputs=X, name='FaceModel')
    
    
    return model
```

```python
FaceModel = model((48, 48, 3))
FaceModel.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
FaceModel.summary()
```

```python
if not os.path.isdir('aug_data/val'):
  os.mkdir('aug_data/val')

for i in NAME:
  if not os.path.isdir('aug_data/val/' + i):
    os.mkdir('aug_data/val/' + i)

for i in NAME:
  src = os.listdir('aug_data/train/' + i)
  move = random.sample(src, 50)
  for f in move:
    shutil.move('aug_data/train/' + i + '/' + f, 'aug_data/val/' + i + '/' + f)
    print "Finish moving {}".format(i)

print "Finish spliting"
```

```python
batch_size = 32
train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory('aug_data/train/',
                                                   target_size=(48, 48),
                                                   batch_size=batch_size,
                                                   class_mode = 'categorical')
val_datagen = ImageDataGenerator(rescale=1./255)
val_generator = val_datagen.flow_from_directory('aug_data/val/',
                                                   target_size=(48, 48),
                                                   batch_size=batch_size,
                                                   class_mode = 'categorical')
```

```python
FaceModel.fit_generator(
        train_generator,
        steps_per_epoch=2239 // batch_size,
        epochs=50,
        validation_data=val_generator,
        validation_steps=250 // batch_size)

model_json = FaceModel.to_json()
with open("face_model_2.json", "w") as json_file:
    json_file.write(model_json)
FaceModel.save_weights('face_model_2.h5')
```

```python
import cv2
import numpy as np
from keras.preprocessing.image import img_to_array
from keras.models import model_from_json
import time
from picamera.array import PiRGBArray
from picamera import PiCamera
```

```python
with open('face_model_2.json', 'rb') as json_file:
    model_json = json_file.read()
    
model = model_from_json(model_json)
model.load_weights('face_model_2.h5')
print "Load model successfully"
```

```python
print "Start video stream"
camera = PiCamera()
camera.resolution = (600, 600)
camera.framerate = 32
cap = PiRGBArray(camera, size=(600, 600))
```

```python
THD = 0.7
for i in camera.capture_continuous(cap, format='bgr', use_video_port=True):
    frame = i.array
    
    image = cv2.resize(frame, (48, 48))
    image = image.astype('float') / 255.0
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)
    
    predict = model.predict(image)[0]
    
    if predict[0] > THD:
        frame = cv2.putText(frame, "Hello, Dr.Luo Yong", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    elif predict[1] > THD:
        frame = cv2.putText(frame, "Hello, Wang Yongjie", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    elif predict[2] > THD:
        frame = cv2.putText(frame, "Hello, Yi Deliang", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    elif predict[3] > THD:
        frame = cv2.putText(frame, "Hello, Zhang Huaizheng", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    elif predict[4] > THD:
        frame = cv2.putText(frame, "Hello, Dr.Zhou Xin", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    else:
        frame = cv2.putText(frame, "Can not recognize", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    
    cv2.imshow("Frame", frame)
    key = cv2.waitKey(1) & 0xFF
    cap.truncate(0)
    
    if key == ord('q'):
        break
```
